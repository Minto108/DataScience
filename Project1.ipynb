```python
import pandas as pd
import numpy as np
```

```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
warnings.filterwarnings('ignore')
```

```python
url = 'https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29'
```

```python
tables = pd.read_html(url)
df = tables[3]
```

```python
df
```

```python
df.shape
```

```python
# Replace column headers with column numbers
df.columns = range(df.shape[1])
```

```python
df
```

```python
# Retain columns with index 0 and 2
df = df[[0, 2]]
```

```python
df
```

```python
# Retain the rows with index 1 to 10
df = df.iloc[1:11]
df
```

```python
df.columns = ['Country', 'GDP (Million USD)']
df
```

```python
df.dtypes
```

```python
df['GDP (Million USD)'] = df['GDP (Million USD)'].astype(int)
df.dtypes
```

```python
df['GDP (Million USD)'] = np.divide(df['GDP (Million USD)'], 1000)
```

```python
df
```

```python
df['GDP (Million USD)'] = np.round(df['GDP (Million USD)']/1000, 2)
df
```

```python
df.rename(columns = {'GDP (Million USD)' : 'GDP (Billion USD)'})
```

```python
df.to_csv('Largest_economies.csv')
```
